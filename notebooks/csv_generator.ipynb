```python
from faker import Faker
import random
import pandas as pd
from datetime import date
from datetime import timedelta, datetime

fake = Faker()

# ----------------------------
# Generate Customers
# ----------------------------

num_customers = 100

system_types = ['Furnace', 'AC', 'Heat Pump', 'Mini Split', 'Not Sure']
system_brands = ['Lennox', 'Trane', 'Carrier', 'Rheem']
apointment_type = ['Service Call', 'Maintenance', 'Free Quote']
cities = ['Vancouver', 'Battle Ground', 'Camas', 'Washougal', 'Ridgefield', 'Woodland', 'Longview', 'Kelso']
start_date = date(2025, 1, 1)
end_date = date(2027, 7, 13)

customers_data = []

for i in range(1, num_customers + 1):
    first_name = fake.first_name()
    last_name = fake.last_name()
    phone = f"({random.randint(200,999)}) {random.randint(200,999)}-{random.randint(1000,9999)}"
    email = fake.email()
    address = f"{random.randint(10,99999)} {fake.street_name()}"
    city = random.choice(cities) 
    state = "WA" 
    zipcode = fake.zipcode()
    system_type = random.choice(system_types)
    system_brand = random.choice(system_brands)
    system_model = f"{random.choice(['X', 'Z', 'Q'])}{random.randint(10000,99999)}"
    install_date = fake.date_between(start_date='-5y', end_date='today')
    system_notes = fake.sentence(nb_words=6)
    last_service_date = fake.date_between(start_date=install_date, end_date='today')
    membership_expire_date = fake.date_between(start_date=start_date, end_date=end_date)

    customers_data.append({
        "customer_id": i,
        "first_name": first_name,
        "last_name": last_name,
        "phone": phone,
        "email": email,
        "address": address,
        "city": city,
        "state": state,
        "zipcode": zipcode,
        "system_type": system_type,
        "system_brand": system_brand,
        "system_model": system_model,
        "install_date": install_date,
        "system_notes": system_notes,
        "last_service_date": last_service_date,
        "membership_expire_date": membership_expire_date
    })

customers_df = pd.DataFrame(customers_data)

# ----------------------------
# Generate Service Calls
# ----------------------------

num_service_calls = 250
servicing_date = date(2020, 1,1)

service_types = ['Service Call', 'Maintenance', 'Free Quote']
statuses = ['Scheduled', 'Completed', 'Canceled']
system_types = ['Furnace', 'AC', 'Heat Pump', 'Mini Split', 'Not Sure']

service_calls_data = []

for i in range(1, num_service_calls + 1):
    customer_id = random.randint(1, num_customers)
    # pick a random date in last 2 years
    random_days = random.randint(0, 730)
    service_date = fake.date_between(start_date=servicing_date, end_date='today')
    type_of_call = random.choice(service_types)
    description = fake.sentence(nb_words=8)
    quoted_price = f"${random.randint(59, 999)}"
    status = random.choice(statuses)

    service_calls_data.append({
        "service_call_id": i,
        "customer_id": customer_id,
        "service_date": service_date,
        "type": type_of_call,
        "description": description,
        "quoted_price": quoted_price,
        "status": status
    })

service_calls_df = pd.DataFrame(service_calls_data)

# ----------------------------
# Save to CSV
# ----------------------------

customers_df.to_csv("fake_customers.csv", index=False)
service_calls_df.to_csv("fake_service_calls.csv", index=False)

print("Synthetic data created and saved!")
```
